```python
import numpy as np
import matplotlib.pyplot as plt  
```

## Create a sample from a three category multinomial distribution 

```python
n = 100 
p = [0.5,0.3, 0.2]
multinomial_sample = np.random.multinomial(n,p)
```

## Visualize the multinomial distribution and the multinomial sample

```python
catergories = ["Catergories 1","Catergories 2","Catergories 3"]
plt.bar(catergories,multinomial_sample, color = ['blue', 'orange', 'green'])
plt.title("Multinomial Sample Distribution")
plt.xlabel("Catergories")
plt.ylabel("Counts")
plt.show()
```

```python
plt.bar(catergories,p, color = ['blue', 'orange', 'green'])
plt.title("Multinomial Probability Distribution")
plt.xlabel("Catergories")
plt.ylabel("Counts")
plt.show()
```

```python
import numpy as np
from scipy import stats
import plotly.figure_factory as ff
```

```python
# p1p2p3_scatter = stats.dirichlet(alpha=np.array([1.0, 1.0, 1.0])).rvs(size=100)
p1 = np.linspace(0,1,100)
p2 = 1-p1.copy()
p_ = np.linspace(0,1,100)
p1 = (p1[np.newaxis,:]*p_[:,np.newaxis]).flatten()
p2 = (p2[np.newaxis,:]*p_[:,np.newaxis]).flatten()
p3 = 1-p1-p2
p = np.c_[p1,p2,p3]

alpha_prior = np.array([1.1, 1.1, 1.1])  # Prior alpha values
dirichlet_prior = stats.dirichlet(alpha=alpha_prior)
dirichlet_prior_densities = dirichlet_prior.pdf(p.T)

# https://plotly.com/python/ternary-contour/Links to an external site. 
fig = ff.create_ternary_contour(p.T, dirichlet_prior.pdf(p.T),
                                pole_labels=['p1', 'p2', 'p3'],
                                interp_mode='cartesian',
                                showscale=True)
fig.show('png')
```
